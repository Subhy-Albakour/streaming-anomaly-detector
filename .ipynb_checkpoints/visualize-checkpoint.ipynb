```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

from cf_tree import CFTree 
from clustering_feature import ClusteringFeature
from geo.point import Point


n_samples = 1500
random_state = 2

X, y= make_blobs(n_samples=n_samples,
                centers=5,
                cluster_std=[1, 5, 15, 20, 25],
                random_state=random_state)

cluster = KMeans(n_clusters=5, random_state=random_state,init="k-means++").fit(X)
y_pred=cluster.predict(X)
```

```python
def bico(coreset_size):

    tree=CFTree(dim=2,thresh=1,max_nodes= coreset_size)

    for row in X:
        tree.insert(Point(row))

    coreset_centers,ws=tree.get_coreset()


    plt.subplot(131)
    plt.scatter(coreset_centers[:,0],coreset_centers[:,1],s=ws)
    plt.title("Reduced data set of size {0}".format(coreset_size))
    
  
    cluster_w=KMeans(n_clusters=5, random_state=random_state,init="k-means++").fit(coreset_centers,sample_weight=ws)
    centroids_w = cluster_w.cluster_centers_
    y_pred_w = cluster_w.predict(X)
    
    plt.subplot(132)
    plt.scatter(centroids_w[:, 0], centroids_w[:, 1],
            marker='o', s=50, linewidths=2,
            color='r', zorder=10)
    plt.scatter(X[:, 0], X[:, 1], c=y_pred_w)
    plt.title("Clustering on reduced data")
    
    centroids = cluster.cluster_centers_
    plt.subplot(133)
    plt.scatter(centroids[:, 0], centroids[:, 1],
        marker='o', s=50, linewidths=2,
        color='r', zorder=10)
    plt.scatter(X[:, 0], X[:, 1], c=y_pred)
    plt.title("Clustering on original data set")
    plt.show()

```

```python
plt.rcParams["figure.figsize"] = [20,4]

for size in [ 20,30, 50, 100]:
    bico(size)
```

```python

import time

n_samples = 30000
random_state = 2

X, y= make_blobs(n_samples=n_samples,
                centers=5,
                cluster_std=[1, 5, 15, 20, 25],
                random_state=random_state)




start1=time.time()
cluster = KMeans(n_clusters=5, random_state=random_state,init="k-means++").fit(X)
y_pred=cluster.predict(X)
end1=time.time()

start3=time.time()
coreset_size=300
tree=CFTree(dim=2,thresh=1,max_nodes= coreset_size)
for row in X:
    tree.insert(Point(row))
coreset_centers,ws=tree.get_coreset()
end3=time.time()

start2=time.time()
cluster_w=KMeans(n_clusters=5, random_state=random_state,init="k-means++").fit(coreset_centers,sample_weight=ws)
y_pred_w = cluster_w.predict(X) 
end2=time.time()

time1=end1-start1
time2=end2-start2
print(time1, time2)
print(end3-start3)
     
```

```python
import data_generator
centers=[(0,0),(5,5),(-5,7)]
n_samples=1000
random_state=3
X,y=data_generator.generate_clusters_with_outliers_2d(centers,n_samples,outlier_ratio=0.1,
                                                      outlier_distance=2,outlier_std_ratio=2)
anoms=(y==1)
non_anoms=(y==0)
cluster = KMeans(n_clusters=3, random_state=random_state,init="k-means++").fit(X[non_anoms])

y_pred=cluster.predict(X[anoms])
plt.scatter(X[anoms,0],X[anoms,1],c=y_pred)
plt.show()
```

```python
dist=cluster.transform(X[anoms])
min_dist=dist[range(len(dist)),y_pred]
plt.plot(min_dist[min_dist>=15],"ro-")
plt.plot(min_dist[min_dist<=15],"go-")
plt.show()
```
